```python
# imports
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import time
from geopy.geocoders import Nominatim
from gmplot import gmplot2
from IPython.core.display import display, HTML
from mpl_toolkits.basemap import Basemap

%matplotlib inline
```

```python
def wp_epl_scrape(team, df, df_columns):
    '''Scrapes data from Washington Post Sports site for English Premier League'''
    
    # instantiate dict & list for temporary storage
    player_dict = {}
    players = []

    # set up link, request website, and create BeautifulSoup objects
    link = 'http://stats.washingtonpost.com/epl/rosters.asp?team='
    link += str(team)
    r = requests.get(link)
    b = BeautifulSoup(r.text, 'lxml')
    
    # locate table and loop through rows to load data to player_dict
    table = b.find_all('table', {'class':'shsTable shsBorderTable'})
    rows = table[0].find_all('tr', {'class':['shsRow0Row','shsRow1Row']})
    for row in rows:
        data = row.find_all('td')
        player_dict['tm'] = b.find('span',{'class':'countryText'}).text
        player_dict['num'] = data[0].text.strip()
        player_dict['player'] = data[1].text.strip()
        player_dict['pos'] = data[2].text.strip()
        player_dict['height'] = data[3].text.strip()
        player_dict['weight'] = data[4].text.strip()
        player_dict['birth_date'] = data[5].text.strip()
        player_dict['birthplace'] = data[6].text.strip()
        players.append(player_dict.copy())

    # create a temporary dataframe with freshly scraped data
    df_temp = pd.DataFrame(players, columns = df_columns)
    # append data to base dataframe
    df = df.append(df_temp, ignore_index=False)
    
    return df

# dictionary of teams with washington post team code
teams = {'Arsenal':21,
         'Bournemouth':1153,
         'Burnley':1133,
         'Chelsea': 25,
         'Crystal Palace':90,
         'Everton':28,
         'Hull City':939,
         'Leicester City':31,
         'Liverpool':32,
         'Manchester City':33,
         'Manchester United':34,
         'Middlesbrough':35,
         'Southampton':37,
         'Stoke City':1139,
         'Sunderland':38,
         'Swansea City':1141,
         'Tottenham Hotspur':39,
         'Watford':97,
         'West Bromwich Albion':406,
         'West Ham United':40}

# create list of columns & base dataframe
cols = ['tm','num','player','pos','height','weight','birth_date','birthplace']
df = pd.DataFrame({}, columns = cols)
# loop through teams to build dataset
for tm, tm_v in teams.items():
    df = wp_epl_scrape(tm_v, df, cols)
    
df.head()
```

```python
# check that the data looks normal
print(df.shape)
print(len(df.tm.unique()))
print(df.isnull().sum())
```

```python
# instantiate geolocator
geolocator = Nominatim()

def f_lat(x):
    try:
        location = geolocator.geocode(x)
    except:
        x = x.split(',')[1]
        location = geolocator.geocode(x)
    try:
        val = location.latitude
        return val
    except:
        pass

# pull geocodes for latitudes
lats = df.birthplace.map(f_lat)
```

```python
def f_lon(x):
    try:
        location = geolocator.geocode(x)
    except:
        x = x.split(',')[1]
        location = geolocator.geocode(x)
    try:
        val = location.longitude
        return val
    except:
        pass

# pull geocodes for longitudes
lons = df.birthplace.map(f_lon)
```

```python
# add latitudes & longitudes to dataframe; re-name columns
df = pd.concat([df, lats, lons], axis=1)
cols.extend(['lat','lon'])
df.columns = cols
```

```python
def hometown_plot(df, team=None):
    '''Creates a world plot of hometowns for the EPL using Basemap library'''
    
    df.dropna(inplace=True)

    if team:
        df = df[df.tm == team]
    
    fig = plt.figure(figsize=(15,15))
    themap = Basemap(projection='gall',
                     llcrnrlon=-180, 
                     llcrnrlat=-60,
                     urcrnrlon=179, 
                     urcrnrlat=85)
    themap.drawcoastlines()
    themap.drawcountries()
    themap.fillcontinents(color = 'gainsboro')
    themap.drawmapboundary(fill_color='steelblue')

    for index, row in df.iterrows():
        x, y = themap(row['lon'], row['lat'])
        themap.plot(x, y, 'o', color='orange', markersize=8)

    plt.title('Hometowns of All EPL Players in 2016', fontsize=20)
    if team:
        plt.title('Hometowns of 2016 {0} Players'.format(team), fontsize=20)
        
    plt.show()

# plot hometowns for all players
hometown_plot(df)
# plot hometown for Chelsea
hometown_plot(df, 'Chelsea')
```
